## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df = pd.read_csv('/Users/miyuwakou/PG/BasicPython/data/hist_data.csv')
plt.hist(df['aデータ'], bins = 30, alpha=0.5, label='aデータ', color="orange")
plt.hist(df['bデータ'], bins = 30, alpha=0.5, label='bデータ', color="turquoise")

plt.xlabel('温度(℃)')
plt.ylabel('度数')
plt.title('aデータとbデータの温度に関する度数分布')
plt.legend()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df2 = pd.read_csv('/Users/miyuwakou/PG/BasicPython/data/scatter_data.csv')

mask0 = df2[df2['class'] == 0]
mask1 = df2[df2['class'] == 1]
mask2 = df2[df2['class'] == 2]

plt.scatter(mask0['length (cm)'], mask0['width (cm)'], c='red', label='class 0', marker='o')
plt.scatter(mask1['length (cm)'], mask1['width (cm)'], c='blue', label='class 1', marker='o')
plt.scatter(mask2['length (cm)'], mask2['width (cm)'], c='green', label='class 2', marker='o')

# x軸のラベルとy軸のラベル
plt.xlabel('length (cm)')
plt.ylabel('width (cm)')

# グラフのタイトル
plt.title('クラス0〜2のlengthとwidthに関する散布図')

# 凡例を表示
plt.legend()
plt.grid(True)

# グラフを表示
plt.show()


```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df3 = pd.read_csv('/Users/miyuwakou/PG/BasicPython/data/line_data.csv')
x = df3['期間']
y = df3['件数']
plt.figure(figsize = (8, 5))

plt.plot(x, y, marker='o')
plt.xlabel('期間')
plt.ylabel('件数')
plt.title('期間ごとの費用の推移')
plt.xticks(rotation=30)
plt.grid(True)

plt.show()


```

```python
df4 = pd.read_csv('/Users/miyuwakou/PG/BasicPython/data/line_data.csv')

x = df4['期間']
y = df4['費用']
plt.figure(figsize = (8, 5))

plt.bar(x, y)

plt.title('期間ごとの費用の推移')
plt.xlabel("期間")
plt.ylabel("費用(円)")
plt.xticks(rotation=30)

plt.grid(True)
plt.show()
```
